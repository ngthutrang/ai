```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
frames = []

for file in glob.glob("*.csv"):
    frames.append(pd.read_csv(file))

df = pd.concat(frames, ignore_index=True)
df
```

```python
algo_map = {'breadth_first_search': 1,
           'depth_first_graph_search': 2,
           'uniform_cost_search': 3,
           'greedy_best_first_graph_search with h_unmet_goals': 4,
           'greedy_best_first_graph_search with h_pg_levelsum': 5,
           'greedy_best_first_graph_search with h_pg_maxlevel': 6,
           'greedy_best_first_graph_search with h_pg_setlevel': 7,
           'astar_search with h_unmet_goals': 8,
           'astar_search with h_pg_levelsum': 9,
           'astar_search with h_pg_maxlevel': 10,
           'astar_search with h_pg_setlevel': 11}

algo_type_map = {'breadth_first_search': 'uninformed',
               'depth_first_graph_search': 'uninformed',
               'uniform_cost_search': 'uninformed',
               'greedy_best_first_graph_search with h_unmet_goals': 'greedy',
               'greedy_best_first_graph_search with h_pg_levelsum': 'greedy',
               'greedy_best_first_graph_search with h_pg_maxlevel': 'greedy',
               'greedy_best_first_graph_search with h_pg_setlevel': 'greedy',
               'astar_search with h_unmet_goals': 'astart',
               'astar_search with h_pg_levelsum': 'astart',
               'astar_search with h_pg_maxlevel': 'astart',
               'astar_search with h_pg_setlevel': 'astart'}
algo_types = set(algo_type_map.values())

df['algo_no'] = df.apply(lambda row: algo_map[row['SearchAlgo']],  axis=1)
df['algo_type'] = df.apply(lambda row: algo_type_map[row['SearchAlgo']],  axis=1)
df
```

```python
for s in algo_types:
    subset = df[df["algo_type"] == s]
    plt.scatter(subset["New_Nodes"], subset["Actions"], label=s)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('New Nodes vs Actions')
plt.xlabel('New_Nodes')
plt.ylabel('Actions')

plt.show()
```

```python
def plot_type(s):
# s = "uninformed"
    subset = df[df["algo_type"] == s]

    for l in algo_map.keys():
        ssubset = subset[subset["SearchAlgo"]==l]
        if len(ssubset)>0:
            plt.scatter(ssubset["New_Nodes"], ssubset["Actions"], label=l)
            plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.title('New Nodes vs Actions {}'.format(s))
    plt.xlabel('New_Nodes')
    plt.ylabel('Actions')
    plt.show()

for s in algo_types:
    plot_type(s)
```

```python
for s in algo_types:
    subset = df[df["algo_type"] == s]
    plt.scatter(subset["Time"], subset["Actions"], label=s)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Time vs Actions')
plt.xlabel('Time')
plt.ylabel('Actions')

plt.show()
```

```python
for s in algo_types:
    subset = df[df["algo_type"] == s]
    plt.scatter(subset["Plan_Length"], subset["Actions"], label=s)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Plan_Length vs Actions')
plt.xlabel('Plan_Length')
plt.ylabel('Actions')

plt.show()
```
